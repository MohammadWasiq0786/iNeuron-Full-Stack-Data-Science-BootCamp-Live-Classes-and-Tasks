## DIFFERENT SAMPLING TECHNIQUES :

1. **Simple Random Sampling :** Every member of the population (N) has an equal chance of being selected for your sample (n) 
<br> Eg. Exit Polls, General Survey, Lottery
2. **Stratified Sampling :** Strata means Layers/Clusters/Groups.
<br> Eg. We can make groups basis Gender, Education Degree etc. Eg. RBI Survey
3. **Systematic Sampling :** (In Airport, Credit Card Section wants to sell credit card.
<br> Every 5th person, I will see, I will approach to sell Credit Card) Basically selecting every nth individual out of Population (N)
4. **Convenience Sampling :** Only those who are interested in the survey will only Participate. 
<br> Eg. Survey Regarding New Technology

We can apply Stratified Sampling first and then on that sample, we can do Random Sampling (Selling Credit Cards)

```python
import numpy.random as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
population_size = 1000000
population = np.rand(1000000)
```

```python
population
```

```python
number_of_samples = 10000
sample_means = np.rand(number_of_samples)
sample_size = 35
```

```python
sample_means
```

```python
import numpy 
c = np.rand(number_of_samples)

for i in range(0,number_of_samples):
    c = np.randint(1,population_size,sample_size)
    sample_means[i] = numpy.mean(c)
```

```python
np.randint(1, population_size, sample_size)
```

```python
sample_means.shape
```

```python
plt.subplot(1,2,1)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.distplot(sample_means, bins=int(180/5), hist = True, kde = False)
plt.title('Histogram of Sample mea', fontsize=20)
plt.xlabel('Sample mean', fontsize=20)
plt.ylabel('Count', fontsize=20)

plt.subplot(1,2,2)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.distplot(sample_means, hist = False, kde = True)
plt.title('Density of Sample mean', fontsize=20)
plt.xlabel('Sample mean', fontsize=20)
plt.ylabel('Density', fontsize=20)
plt.subplots_adjust(bottom=0.1, right=2, top=0.9)
```
