```python
def outlier_threshold(df, variable):
    q1= df[variable].quantile(0.25)
    q3= df[variable].quantile(0.75)
    iqr= q3-q1
    up_limit= q3+(1.5*iqr)
    lower_limit= q1-(1.5*iqr)
    return lower_limit, up_limit   
```
