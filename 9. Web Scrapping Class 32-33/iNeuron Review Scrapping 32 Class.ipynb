````python
from flask import Flask, render_template, request, jsonify
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as urReq
````

````python
flipkart_url = "https://www.flipkart.com/search?q="+"iphone11"
````

````python
flipkart_url
````

````python
response_website = urReq(flipkart_url) #Through this URL Request, we are able to hit this URL using Python
````

````python
data_flipkart = response_website.read() #This information is same as View Page Source of Flipkart page written in HTML
data_flipkart
````

````python
beautifyed_html = bs(data_flipkart,"html.parser") #Some indentation change: Beautifying information
beautifyed_html #Now we have all the information on Flipkart Iphone 11 page
````

````python
bigbox = beautifyed_html.find_all("div",{"class":"_1AtVbE col-12-12"}) #Go to Inspect Element of the URL to understand this process
bigbox[0]
````

````python
bigbox[6].div.div.div.a['href'] #For 6th Box, we were able to find the URL
````
